```python
import cv2

## CALLBACK FUNCTION ---> DRAWING A RECTANGLE
## HOW TO ACTUALLY UPDATE PT1 AND PT2

def draw_rectangle(event, x, y, flags, params):
    
    global pt1, pt2, topLeft_clicked, botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        # THE RECTANGLE HAS ALREADY BEEN DRAWN
        # RESET THE RECTANGLE (DELETE THE RECTANGLE)
        if topLeft_clicked == True and botRight_clicked == True:
            pt1 = (0,0)
            pt2 = (0,0)
            topLeft_clicked = False
            botRight_clicked = False
            
        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True
        
        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True
         
    

## GLOBAL VARIABLES

pt1 = (0,0) # TOP LEFT
pt2 = (0,0) # BUTTOM RIGHT

# THEY INDICATE WHETHERE OR NOT I CLICKED THE POINT THAT I INTED TO BE TOP LEFT AND BUTTOM RIGHT
# AS THEY GET CLICKED I UPDATE POINT ONE A ND POINT 2 
# IF BOTH ARE TRUE IT MEANS I HAVE DRAWN THE RECTANGLE
topLeft_clicked = False
botRight_clicked = False


## CONNECT TO CALLBACK

cap = cv2.VideoCapture(0)

cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_rectangle)

while True:
    
    ret, frame = cap.read()
    
    # DRAWING ON THE FRAME BASED OFF THE GLOBAL VARIABLES
    
    if topLeft_clicked == True:
        cv2.circle(frame, center=pt1, radius=5, color=(0,0,255), thickness=-1)
    
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0,0,255),3)
   
    
    cv2.imshow('Test', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
