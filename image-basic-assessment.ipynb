### Open the dog_backpack.jpg image from DATA folder and display it in the notebook.
### correct the RGB order

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('DATA/dog_backpack.jpg') # ---> array

# Matplotlib: RGB
# OpenCV: BGR

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Flip the image upside down and display it in the notebook

```python
new_img = cv2.flip(img, 0) # ----> 0: x axis, 1: y axis
plt.imshow(new_img)
```

### Draw an empty rectangle around the dog's face and display the image in the notebook

```python
cv2.rectangle(img, (200,400), (600,800), color=(255,0,0), thickness=5)
plt.imshow(img)
```

### Draw a BLUE TRIANGLE in the middle of the image. The size and the angle is up to you, but it should be a triangle in any orientation

```python
img = cv2.imread('DATA/dog_backpack.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

vertices = np.array([[480,400],[200,800],[600,800]], dtype=np.int32)

pts = vertices.reshape((-1,1,2))

cv2.polylines(img, [pts], isClosed=True, color=(0,0,255), thickness=4)

plt.imshow(img)
```

```python
### Fill in the triangle

img = cv2.imread('DATA/dog_backpack.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

vertices = np.array([[480,400],[200,800],[600,800]], dtype=np.int32)

pts = vertices.reshape((-1,1,2))

cv2.fillPoly(img, pts=[pts], color=(0, 0, 255))

plt.imshow(img)
```

### (Note: You will need to run this as a script), Create a script that opens the picture and allows you to draw empty red circles whereever you click the RIGHT MOUSE BUTTON DOWN.

```python
# Callback Function

def draw_circle(event, x, y, flags, params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x,y), 100, (0,0,255), thickness=10)
        
        
img = cv2.imread("DATA/dog_backpack.png")

cv2.namedWindow(winname ='dog')

cv2.setMouseCallback('dog', draw_circle)

while True:
    cv2.imshow('dog', img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
