```python
import cv2


# connect to a camera

cap = cv2.VideoCapture(0) # 0 means it's gonna grab your default camera

# it's better to garb the width and height of the video frame
# it's gonna return a floating point number: 1080.0
# it's better to cast it as integers
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


writer = cv2.VideoWriter('mysupervideo.mp4', cv2.VideoWriter_fourcc(*'XVID'), 20, (width, height))
while True:
    
    # reading the image files(frames)
    ret, frame = cap.read()
    
    
    # operations
    
    writer.write(frame)
    
    # converting frames to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    cv2.imshow('frame', gray)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
        
# stop capturing the frames        
cap.release()
writer.release()
cv2.destroyAllWindows()
```
