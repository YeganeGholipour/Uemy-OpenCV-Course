```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
full = cv2.imread('DATA/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    # CREATE A COPY OF THE IMAGE
    full_copy = full.copy()
    
    method = eval(m)
    
    # TEMPLATE MATCHING
    
    res = cv2.matchTemplate(full_copy, face, method)
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res) # it looks at the heatmap and looks for the min and maximum location
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]: # for these two methods the minimum location is where it has detected the object 
        top_left = min_loc  #(x, y)  # (width, height)
    else:
        top_left = max_loc   # top_left is where we have first detected the object
    
    
    height, width, channels = face.shape
    
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(full_copy, top_left, bottom_right, (255,0,0), 10)
    
    # PLOT AND SHOW THE IMAGE
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEATMAP OF TEMPLATE MATCHING')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('HEATMAP OF TEMPLATE')
    # TITLE WITH THE METHOD USED
    plt.suptitle(m)
    
    plt.show()
    
    print('\n')
    print('\n')
```
